```python
import os
import sys
import numpy as np
import pandas as pd

from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors
from scipy.stats import chi


# -----------------------------
# Create output folders
# -----------------------------

main_out = "/define/main/parent/directory/"
out_dir = os.path.join(main_out, "sub_directory_for_different_scenarios")

os.makedirs(out_dir, exist_ok=True)
print(f"out_dir is: {out_dir}")


# -----------------------------
# Read historical and projected data
# -----------------------------

Historical = pd.read_csv("/data/Amatrix_gridmet.csv")
HistoricalICV = pd.read_csv("/data/Cmatrix_gridmet.csv")
Projected = pd.read_csv("/data/Bmatrix.csv")


# -----------------------------
# Define variables
# -----------------------------

time = "2040_70"
emission = "RCP85"

models = Projected["model"].unique()
a_model = models[0]

Projected = Projected[Projected["model"] == a_model].copy()

# Drop the model column if needed
if "model" in Projected.columns:
    Projected = Projected.drop(columns=["model"])


# -----------------------------
# Select columns used for analogs
# R equivalent:
# HistoricalICV(1:2,5,13:16)
# In Python, column positions are 0-based
# -----------------------------

selected_col_idx = [0, 1, 4, 12, 13, 14, 15]

Historical = Historical.iloc[:, selected_col_idx].copy()
Projected = Projected.iloc[:, selected_col_idx].copy()
HistoricalICV = HistoricalICV.iloc[:, selected_col_idx].copy()


# -----------------------------
# Define nonnumeric and numeric columns
# -----------------------------

non_numeric_cols = Historical.columns[:2].tolist()
numeric_cols = Historical.columns[2:7].tolist()

local_locations = Projected["State_County"].unique()

trunc_SDs = 0.1
n_neighbors = 5  # change this if needed


# -----------------------------
# Read command line argument
# Example:
# python script.py 1
# -----------------------------

args = sys.argv[1:]

if len(args) > 0:
    loc_index = int(args[0]) - 1   # R is 1-based; Python is 0-based
else:
    loc_index = 0

a_future_loc = local_locations[loc_index]

print(args)
print(a_future_loc)


# -----------------------------
# Filter projected and reference data
# -----------------------------

Bj = Projected[Projected["State_County"] == a_future_loc].copy()
Cj = HistoricalICV[HistoricalICV["State_County"] == a_future_loc].copy()

print(Cj.shape)


# -----------------------------
# Step 1:
# Express climate data as standardized anomalies
# of the reference period
# -----------------------------

Cj_sd = Cj[numeric_cols].std(skipna=True)
Cj_sd[Cj_sd < 1e-10] = 1

A_prime = Historical.copy()
A_prime[numeric_cols] = A_prime[numeric_cols] / Cj_sd

Bj_prime = Bj.copy()
Bj_prime[numeric_cols] = Bj_prime[numeric_cols] / Cj_sd

Cj_prime = Cj.copy()
Cj_prime[numeric_cols] = Cj_prime[numeric_cols] / Cj_sd


# -----------------------------
# Step 2:
# PCA on reference-period ICV
# and project all data
# -----------------------------

Cj_pca_input = Cj_prime[numeric_cols].dropna()

pca = PCA()
pca.fit(Cj_pca_input)

explained_sd = np.sqrt(pca.explained_variance_)

PCs = np.max(np.where(explained_sd > trunc_SDs)[0]) + 1

X_scores = pca.transform(A_prime[numeric_cols])
Yj_scores = pca.transform(Bj_prime[numeric_cols])
Zj_scores = pca.transform(Cj_prime[numeric_cols])

pc_names = [f"PC{i+1}" for i in range(X_scores.shape[1])]

X = pd.concat(
    [
        A_prime[non_numeric_cols].reset_index(drop=True),
        pd.DataFrame(X_scores, columns=pc_names)
    ],
    axis=1
)

Yj = pd.concat(
    [
        Bj_prime[non_numeric_cols].reset_index(drop=True),
        pd.DataFrame(Yj_scores, columns=pc_names)
    ],
    axis=1
)

Zj = pd.concat(
    [
        Cj_prime[non_numeric_cols].reset_index(drop=True),
        pd.DataFrame(Zj_scores, columns=pc_names)
    ],
    axis=1
)


# -----------------------------
# Step 3a:
# Express PC scores as standardized anomalies
# of reference interannual variability
# -----------------------------

selected_pcs = pc_names[:PCs]

Zj_sd = Zj[selected_pcs].std(skipna=True)
Zj_sd[Zj_sd < 1e-10] = 1

X_prime_numeric = X[selected_pcs] / Zj_sd
Yj_prime_numeric = Yj[selected_pcs] / Zj_sd

X_prime = pd.concat(
    [X[non_numeric_cols].reset_index(drop=True),
     X_prime_numeric.reset_index(drop=True)],
    axis=1
)

Yj_prime = pd.concat(
    [Yj[non_numeric_cols].reset_index(drop=True),
     Yj_prime_numeric.reset_index(drop=True)],
    axis=1
)

X_prime = X_prime.replace([np.inf, -np.inf], np.nan).dropna()
Yj_prime = Yj_prime.replace([np.inf, -np.inf], np.nan).dropna()


# -----------------------------
# Step 3b:
# Nearest neighbors
# -----------------------------

nn = NearestNeighbors(
    n_neighbors=n_neighbors,
    algorithm="brute",
    metric="euclidean"
)

nn.fit(X_prime[selected_pcs])

NN_dist, NN_idx = nn.kneighbors(Yj_prime[selected_pcs])


# -----------------------------
# Convert distance to chi percentile
# then to sigma values
# -----------------------------

NN_chi = chi.cdf(NN_dist, df=PCs)
NN_sigma = chi.ppf(NN_chi, df=1)


# -----------------------------
# Build output tables
# -----------------------------

NN_dist_tb = pd.DataFrame(
    NN_dist,
    columns=[f"NN_{i+1}" for i in range(n_neighbors)]
)

NN_dist_tb.insert(
    0,
    "State_County",
    Yj_prime["State_County"].values
)

NN_sigma_tb = pd.DataFrame(
    NN_sigma,
    columns=[f"sigma_NN_{i+1}" for i in range(n_neighbors)]
)

NN_sigma_tb.insert(
    0,
    "State_County",
    Yj_prime["State_County"].values
)

neighbor_locations = X_prime.iloc[NN_idx.flatten()]["State_County"].values
neighbor_locations = neighbor_locations.reshape(NN_idx.shape)

NNs_loc_year_tb = pd.DataFrame(
    neighbor_locations,
    columns=[f"Loc_NN_{i+1}" for i in range(n_neighbors)]
)

NNs_loc_year_tb.insert(
    0,
    "State_County",
    Yj_prime["State_County"].values
)


# -----------------------------
# Combine neighbor locations and sigma values
# Similar to final R melt output
# -----------------------------

wide_out = pd.DataFrame()
wide_out["State"] = NNs_loc_year_tb["State_County"]

for i in range(n_neighbors):
    wide_out[f"N{i+1}"] = NNs_loc_year_tb[f"Loc_NN_{i+1}"]
    wide_out[f"V{i+1}"] = NN_sigma_tb[f"sigma_NN_{i+1}"]

long_out = wide_out.melt(
    id_vars="State",
    var_name="variable",
    value_name="value"
)


# -----------------------------
# Write output
# -----------------------------

safe_model = str(a_model).replace("-", "_")

output_file = os.path.join(
    out_dir,
    f"NN_loc_year_tb_{a_future_loc}_{safe_model}_{time}_{emission}.csv"
)

long_out.to_csv(output_file, index=False)

print(f"Saved: {output_file}")
```
